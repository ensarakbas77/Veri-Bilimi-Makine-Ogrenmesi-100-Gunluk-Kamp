```python
age_list = [10,20,30,30,30,40,60,70,75]
weight_list = [20,60,80,85,86,87,70,90,90]
```

```python
plt.plot(age_list, weight_list, "g")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
# plt.savefig("test.png")
```

```python
# with numpy
x = np.linspace(0,10,20)
x
```

```python
y = x ** 2
y
```

```python
plt.plot(x,y,"r") # --> Also we can use r*, r*-, r-- 
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.1,0.1,0.6,0.6])
figure_axes.plot(x,y,"g--")
figure_axes.set_xlabel("X Axis")
figure_axes.set_ylabel("Y Axis")
figure_axes.set_title("Graph Title")
plt.show()
```

```python
new_fig = plt.figure(dpi=100) # dpi --> çözünürlük
new_axes = new_fig.add_axes([0.1,0.1,0.9,0.9])
new_axes.plot(x, x ** 2, label = "x ** 2")
new_axes.plot(x, x ** 3, label = "x ** 3")
new_axes.legend(loc=2)
plt.show()
```

# Styles

```python
data1 = np.linspace(0,10,20)
data1
```

```python
data2 = data1 ** 2
data2 
```

```python
my_fig, my_axes = plt.subplots()
print(type(my_fig))
print(type(my_axes))
print()
my_axes.plot(data1, data2, color = "#1E0703", alpha= 0.5)
my_axes.plot(data2, data1, color = "#53CA0A")
plt.show()
```

```python
(new_figure, newAxes) = plt.subplots()

newAxes.plot(data1, data1 + 2, color="blue", linewidth = 1.0)
newAxes.plot(data1, data1 + 3, color="yellow", linewidth = 4.0)
newAxes.plot(data1, data1 + 4, color="#c85d09", linestyle = "-.")
newAxes.plot(data1, data1 + 5, color="green", linestyle = ":" , linewidth = 2.0)
newAxes.plot(data1, data1 + 6, color="#000000", linestyle = "--" , marker="o", markersize=4, markerfacecolor="red")
plt.show()
```
