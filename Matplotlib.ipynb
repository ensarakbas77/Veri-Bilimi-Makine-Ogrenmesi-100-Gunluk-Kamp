```python
age_list = [10,20,30,30,30,40,60,70,75]
weight_list = [20,60,80,85,86,87,70,90,90]
```

```python
plt.plot(age_list, weight_list, "g")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
# plt.savefig("test.png")
```
