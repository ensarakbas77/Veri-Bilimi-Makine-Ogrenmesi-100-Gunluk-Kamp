```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Olimpiyat Verisini Anlamak

```python
df = pd.read_csv('athlete_events.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
print(df.columns)
print()
print(df["ID"].count())
print()
print(df.shape)
```

```python
plt.scatter("Height", "Weight", data=df)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Athelete Weight vs Height")
plt.show()
```

```python
# with seaborn
sns.set_style("whitegrid") # --> white, whitegrid, dark, darkgrid
sns.scatterplot(x="Height", y="Weight",data=df)
plt.xlabel("Height of Atheletes")
plt.ylabel("Wight of Atheletes")
plt.title("Atheletes Weight vs Height")
plt.show()
```

```python
# male vs female
sns.set_style("darkgrid") 
sns.scatterplot(x="Height", y="Weight", hue="Sex", data=df)
plt.xlabel("Height of Atheletes")
plt.ylabel("Wight of Atheletes")
plt.title("Atheletes Weight vs Height")
plt.show()
```

# Graphic Types

```python
df["Medal"].unique()
```

```python
sns.set_style("white")
sns.lineplot(x="Height", y="Weight", hue="Sex", data=df)
plt.xlabel("Height of Atheletes")
plt.ylabel("Wight of Atheletes")
plt.title("Atheletes Weight vs Height")
plt.show()
```

```python
# Height Distributin
sns.set_style("white")
sns.displot(x="Height", hue="Sex", kind="kde", data=df) # kde ***
plt.ylabel("Frequency")
plt.title("Atheletes Height Distributin")
plt.show()
```

```python
sns.barplot(x="Medal", y="Height", hue="Sex", data=df)
plt.title("Medal by Height")
plt.show()
```

```python
sns.catplot(x="Medal", y="Height", hue="Sex", col="Season", data=df)
```

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
```

```python
# box plot --> median, quartile(%25, %50, %75), min max * 1.5 IQR, outlier


np_data = np.array([5,7,9,15,20,22,25,30,32,35,37,40,50,55,60,100]) # 100 is outlier data

plt.figure(figsize=(6,5))
sns.boxplot(y=np_data)
plt.title("Box Plot")
plt.ylabel("Data Value")
#plt.grid(True)
plt.show()
```

```python
# TITANIC
df_titanic = sns.load_dataset("titanic")
df_titanic
```

```python
plt.figure(figsize=(13,6))

plt.subplot(1,2,1)
sns.boxplot(x="class", y="age", data=df_titanic)
plt.title("Age by Class")
plt.xlabel("Class")
plt.ylabel("Age")

plt.subplot(1,2,2)
sns.boxplot(x="class", y="fare", data=df_titanic)
plt.title("Fare by Class")
plt.xlabel("Class")
plt.ylabel("Fare")

plt.tight_layout()
plt.show()
```
